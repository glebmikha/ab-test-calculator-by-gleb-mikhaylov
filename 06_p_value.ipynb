```python
import pandas as pd
import numpy as np
```

# P-value for 7/10

```python
np.random.binomial(10,0.5)
```

```python
n = 10000
result = []
for i in range(n):
  np.random.seed(i)
  result.append(np.random.binomial(10,0.5))
```

```python
df = pd.DataFrame(result)
```

```python
df
```

```python
df['dev'] = abs(df[0] - 5)
```

```python
df
```

```python
(df['dev'] >= 2).mean()
```

# P-value scale

```python
devs = [0,1,2,3,4,5]
```

```python
result = []
for dev in devs:
  result.append((dev,(df['dev'] >= dev).mean()))
```

```python
pvals = pd.DataFrame(result,columns=['heads','p-values'])
```

```python
pvals
```

# P-value for satisfaction case

## P-value 84 / 115

```python
sample_size = 115
```

```python
n = 10000
result = []
for i in range(n):
  np.random.seed(i)
  result.append(np.random.binomial(sample_size,0.8))
```

```python
df = pd.DataFrame(result)
```

```python
sample_size * 0.8
```

```python
df['dev'] = df[0] - sample_size * 0.8
```

```python
df
```

```python
(df['dev'] <= -8).mean()
```

## Scale

```python
devs = list(range(0,int(sample_size * 0.8) + 1))
```

```python
result = []
for dev in devs:
  result.append((dev,(df['dev'] <= -dev).mean()))
```

```python
pvals = pd.DataFrame(result,columns=['people','p-values'])
```

```python
pvals
```

```python
pvals.set_index('people').plot()
```

```python
pvals.head(18)
```

# P-value with statistical test

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
proportions_ztest(84,115,0.8,alternative='smaller')
```

```python
pvals['pval_ztest'] = pvals['people'].apply(lambda x: proportions_ztest(92-x,115,0.8,alternative='smaller')[1])
```

```python
pvals
```

```python
pvals.set_index('people').plot()
```
